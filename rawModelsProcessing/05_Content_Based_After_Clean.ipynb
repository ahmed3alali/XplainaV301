# Notebook 5 — Content-Based Filtering (After Adding Predicted Courses)

This notebook re-evaluates content-based filtering after building the completed ratings dataset.

```python
from pathlib import Path
import sys
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pickle
plt.rcParams.update({'figure.dpi': 120, 'axes.spines.top': False, 'axes.spines.right': False})

PROJECT_ROOT = Path('..').resolve()
DATA_RAW = PROJECT_ROOT / 'data' / 'raw'
DATA_PROCESSED = PROJECT_ROOT / 'data' / 'processed'
FIGURES_DIR = PROJECT_ROOT / 'figures'
MODELS_DIR = PROJECT_ROOT / 'models'
RESULTS_DIR = PROJECT_ROOT / 'results'

sys.path.append(str(PROJECT_ROOT))
import utils_recommender as ur
ur.ensure_dirs(FIGURES_DIR, MODELS_DIR, RESULTS_DIR)
```

```python
ratings_full = pd.read_csv(DATA_PROCESSED / 'ratings_full_with_predictions.csv')
courses = pd.read_csv(DATA_PROCESSED / 'final_courses.csv')

content_assets = ur.create_content_similarity(courses)
tfidf_cos_df = content_assets['tfidf_similarity']
bow_cos_df = content_assets['bow_similarity']
```

```python
with open(MODELS_DIR / 'tfidf_similarity_after.pkl', 'wb') as f:
    pickle.dump(tfidf_cos_df, f)
with open(MODELS_DIR / 'bow_similarity_after.pkl', 'wb') as f:
    pickle.dump(bow_cos_df, f)

print('Similarity matrices saved.')
```

```python
eval_tfidf_after = ur.evaluate_content_model(tfidf_cos_df, ratings_full, n_users=300)
eval_bow_after = ur.evaluate_content_model(bow_cos_df, ratings_full, n_users=300)

results_cb_after = pd.DataFrame([eval_tfidf_after, eval_bow_after], index=['TF-IDF', 'BoW']).round(4)
results_cb_after
```

```python
results_cb_after.to_csv(RESULTS_DIR / 'results_content_based_after.csv')
print('Saved:', RESULTS_DIR / 'results_content_based_after.csv')
```
